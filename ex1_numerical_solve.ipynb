### Packages we need

```python
#Import Packages
import sympy as sy 
#this is a very handy library that deals with symbolic algebra. Good to find analytical solutions.
import sympy.plotting as syplt
```

### Initializing x as a symbolic variable

This is useful in order generate a function of the symbolic variable and then plot it.

```python
x = sy.Symbol('x')
```

### Creating function with sympy

With this package we just need to assign the function directly

```python
f = 0.5*x**-0.5+0.3*x**-0.2
'''Then for Q=1 we can find the function that of 1 - f. 
When this function hits zero, we find the price that clears''' 
f_comp = 1 - f
f
```

### Lets plot it

Use the syplt.plot

```python
%config InlineBackend.figure_format = 'retina'
# this last magic function can be useful in order to increase the plot resolution in retina displays.
# If you want to use it just uncomment the first line in this cell.
p = syplt.plot(f) 
p2 = syplt.plot(f_comp)
p.append(p2[0])
```

### Evaluating expressions

```python
#lets evaluate both functions at x=0.5
f.subs(x,0.566), f_comp.subs(x,0.566)
```

Lets take a closer look by limiting the domain of the function from 0 to 4

```python
p = syplt.plot(f, xlim=[0,4], ylim=[0,6]) 
```

```python
f_comp.subs(x,0.5671)
# p=0.5671 is a good aproximation for a price that clears the demand when Q=1.
```

As you can see, the excess demand is essentialy zero.

```python
y=x
%config InlineBackend.figure_format = 'retina'

p = syplt.plot(x-f_comp,y, xlim=[0,4], ylim=[0,1.5], show=False)
p[0].line_color = 'green'
p[1].line_color = 'red'
p.show()
```

```python
def fun(p):
    return 0.5*p**-0.5+0.3*p**-0.2

fun(0.5)
```

```python
def ifun(p):
    return 1-fun(p)

ifun(0.5671)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.05,5,0.1)
plt.plot(x, fun(x))
plt.title('Original Demand Function')
plt.show()
```

```python
x = np.arange(0.1,5,0.1)
plt.plot(x, ifun(x))
plt.plot(x, x*0)
plt.title('Inverse Demand Function Q=1')
plt.show()
x*0
```
