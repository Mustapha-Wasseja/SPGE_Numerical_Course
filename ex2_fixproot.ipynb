# Exercise 2 - Fixed Point Iteration

```python
def fixproot(f,x0,tol,maxiter):
    """
    Takes as inputs a function, a guess for the root x0 and searches for 
    the root of f with tol precision and at most in maxiter iterations. 
    """

    gn = lambda x: x-f(x)
    iter = 0
    d=1
    while (abs(d) > tol and iter < maxiter):
        x = gn(x0) 
        d=abs(x-x0) 
        x0=x
        iter += 1
    print('The number of iterations is ' + str(iter))
    
    if iter == maxiter:
        print('maximum number of iterations reached, convergenge not assured')
    
    return x0
```

```python
def fixproot(f,x0,tol,maxiter):
    gn = lambda x: x-f(x)
    iter = 0
    while (abs(x0-gn(x0)) > tol and iter < maxiter):
        x0 = gn(x0)
    return(x0)
    
```

```python
#Import Packages
import sympy as sy 
#this is a very handy library that deals with symbolic algebra. Good to find analytical solutions.
import sympy.plotting as syplt


x = sy.Symbol('x')

f = 0.5*x**-0.5+0.3*x**-0.2
f_comp = 1 - f

y=x
%config InlineBackend.figure_format = 'retina'

p = syplt.plot(x-f_comp,y, xlim=[0,4], ylim=[0,1.5], show=False)
p[0].line_color = 'green'
p[1].line_color = 'red'
p.show()
```
